# A2.0. The XY Model and The Ising Model in One and Two Dimensions

### Introduction

We discuss the physics of the classical paramagnet and the Ising model in one and two dimensions. The Ising Model is a mathematical model in statistical mechanics, which can describe the phenomenon of ferromagnetism. The model consists of discrete varaibles that represent magnetic dipole moments of atomic spins that can be in one of two states. The spins are arranged in a lattice, allowing each spin to interact with its neighbors. This allows for identification of phase transitions, as a simplified model of reality. The two-dimensional square lattice Ising model is one of the simplest statistical models to show a phase transition.

The XY model is an important spin model of easy plan anisotropic magnetic insulators where the spins (which have two components) reside on a set of lattice points.  We take the lattice to be a three-dimensional cubic lattice.

### Analytical Solution of the XY Paramagnet

Consider a collection of $N$ independent XY spins $\bold S_i = (S_{xi}, S_{yi}, 0)$ in an external magnetic field $\bold h$, such that $\bold h \cdot \bold S_i = hS \cos(\phi_i)$ where $\phi_i$ is the angle between the spin and the external magnetic field and $S$ is the fixed length of the spin. We set $J=0$ such that all spins are independent.

### Canonical Partition Function $Z$

To determine thermodynamic quantities, we calculate the partition function $Z$ is given $Z=\sum_{i=1} e^{-\beta E_i}$ where $\beta$ is a so-called Lagrange multiplier and found to be $\beta=1/k_bT$. In classical mechanics, the variable $i$ often becomes continuous and the sum is replaced by an integral. In our case, $\phi\in [0,2\pi]$ is a continuous variable, thus inserting our expression for $H$ into our expression for $Z$, we get $Z_1=2\pi I_0(\beta h)$. Here, $I_0$ is the modified Bessel function of the first kind and order zero.

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.special as sci
import time
import itertools as it
from matplotlib import rc
```

Here, we plot the modified Bessel function of orders 0 through 3.

```python
plt.figure(figsize=(16, 4))
x = np.linspace(0, 4, 100)
plt.plot(x, sci.i0(x))
plt.plot(x, sci.iv(1, x))
plt.plot(x, sci.iv(2, x))
plt.plot(x, sci.iv(3, x))

plt.title('Modified Bessel Function')
plt.xlabel('$x$')
plt.ylabel('$I_v$')
plt.legend(['$I_0$', '$I_1$', '$I_2$', '$I_3$'], loc=2)

plt.show()
```

### Average Magnetization per Spin $m$

Knowing the partition function $Z$, we calculate the component of the average magnetization that is parallel to the external field and the specific heat of the system. This is given $m=\frac1N\lang\hat{h}\cdot\bold M\rangle=I_1(\beta h)/I_0(\beta h)$.

```python
plt.figure(figsize=(16,4))
beta_hs = np.linspace(0, 16, 200)
plt.plot(beta_hs, sci.i1(beta_hs)/sci.i0(beta_hs))
plt.title('Magnetization per Spin in the XY Model')
plt.xlabel('$\\beta h$')
plt.ylabel('$m$')
plt.show()
```

Here we see that the magnetization per spin goes to zero when $\beta h$ goes to zero (when $h\rightarrow0$ or $T\rightarrow\infty$). The first limit means the average magnetization vanishes when we remove the external magnetic field. The latter limit means that as the temperature rises, the thermal fluctuations will begin to dominate the aligning effect of the external field.
